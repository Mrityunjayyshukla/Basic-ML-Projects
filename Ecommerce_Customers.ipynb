```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
data = pd.read_csv("Ecommerce_Customers.csv")
```

```python
data.drop("Address", axis=1, inplace=True)
data.drop("Avatar", axis=1, inplace=True)
# Drops the Address and Avatar column from the section
```

```python
# Setting the features and label
X = data.drop("Yearly Amount Spent", axis=1)
y = data['Yearly Amount Spent']
```

```python
type(y)
```

```python
#Implementing Feature Encoding and Scaling
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X['Email']=le.fit_transform(X['Email'])
```

```python
# Splitting the training and testing Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1234, train_size=0.8)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# Implementing Linear Regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_pred, y_test)
```
